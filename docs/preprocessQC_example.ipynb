```python
merged_values.rename(columns={'Address':'CtrlAddress'},inplace=True)
merged_values['CtrlAddress'] = merged_values['CtrlAddress'].astype(str)
```

```python
allmerged = pd.merge(ctrls,merged_values,how='outer',on='CtrlAddress')
allmerged.head()
```

```python
allmerged.shape
```

```python
allmerged.drop(columns=['AddressB','Type','Color','Genome_Build','CHR','MAPINFO','Strand',
                       'TypeFlag'],inplace=True)
allmerged.head()
```

```python
control_data = pd.DataFrame(columns=['CtrlAddress','Name','AddressA','Sample','idat_color','Value'])
for sample in allmerged.columns:
    if sample not in ['CtrlAddress','Name','AddressA','Sample','idat_color']:
        df = allmerged[['CtrlAddress','Name','AddressA']]
        df['Sample'] = sample
        df['idat_color'] = allmerged['idat_color']
        df['Value'] = allmerged[sample]
        control_data = pd.concat([control_data,df])
control_data.shape
```

```python
control_data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# green staining
greenstain = control_data[
    (control_data['Name']=='STAINING')
    &(control_data['AddressA']!='-99')
    &(control_data['idat_color']=='Green')]
redstain = control_data[
    (control_data['Name']=='STAINING')
    &(control_data['AddressA']!='-99')
    &(control_data['idat_color']=='Red')]

fig = plt.figure(figsize=(10,8))
left = fig.add_subplot('121')
left = sns.stripplot(x="Sample", y="Value", hue="AddressA",size=8,data=greenstain,jitter=True,palette=greenstain['AddressA'].values)
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Staining Green')
plt.xlabel('Sample')
plt.ylabel('Intensity')

right = fig.add_subplot('122')
right = sns.stripplot(x="Sample", y="Value", hue="AddressA",size=8,data=redstain,jitter=True,palette=redstain['AddressA'].values)
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title('Staining Red')
plt.xlabel('Sample')
plt.ylabel('Intensity')
plt.suptitle('Staining Controls').set_y(1.05)
plt.tight_layout()
```
